```python
df_spark_parquet = spark.read \
    .parquet('yellow_tripdata_2021-01.parquet')
```

```python
df_partition_spark = spark.read.parquet('partition/2021/01/')
```

```python
df_partition_spark.printSchema()
```

```python
df_partition_spark.select('tpep_pickup_datetime', 'tpep_dropoff_datetime', 'PULocationID', 'DOLocationID')\
    .filter (df_partition_spark.PULocationID == 1) \
    .show()
```

```python
from pyspark.sql import functions as F
df_partition_spark.withColumn('new', F.to_date(df_partition_spark.tpep_dropoff_datetime)) \
    .show()
```
